#### CDR --> Clinical Dementia Rating

```python
#Como el atributo Group es categorico y este est como tipo object, usamos el replace para asignarle un valor de tipo int
dataset.Group = dataset.Group.replace('Nondemented', 0)
dataset.Group = dataset.Group.replace('Demented', 1)
dataset.Group = dataset.Group.replace('Converted', 2)
```

```python
dataset.head()
```

```python
y = dataset["Group"]
X = dataset.drop("Group",axis=1)
```

```python
dataset.head()
```

## 3. Rellenamos los datos vacios.

### En las columnas SES y MMSE encontramos datos vacios, usamos un imputador del Scikitlearn para llenar estos datos.

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
```

```python
dataset_num = X.drop("M/F", axis=1)
imputer.fit(dataset_num) #Entrenamos el imputador
dataset_temp = imputer.transform(dataset_num) #Aplicamos el imputador y realizamos cambios
```

```python
dataset_temp = pd.DataFrame(dataset_temp, columns=dataset_num.columns, index=dataset_num.index)
X["SES"] = dataset_temp["SES"]
X["MMSE"] = dataset_temp["MMSE"]
```

```python
dataset
```

```python
#Comprobamos los cambios realizados.
X.info()
```

## 5. Aplicamos normalizacin

```python
from sklearn.preprocessing import StandardScaler

dataset_num = X.drop("M/F", axis=1)

scaler = StandardScaler()
alzheimer_scaler = scaler.fit_transform(dataset_num)
alzheimer_scaler
```

```python
X_scaler = pd.DataFrame(alzheimer_scaler, columns=dataset_num.columns, index=dataset_num.index)
X_scaler
```

```python
X[dataset_num.columns] = alzheimer_scaler
X
```

```python
## 6. Juntamos la columna M/F con el resto de la tabla.
```

```python
X = pd.get_dummies(X)
X
```

## 7. Separamos los datos en dos grupos: Train y Test.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
print("X_train:", X_train.shape,"y_train: ", y_train.shape)
print("X_test:", X_test.shape, "y_test", y_test.shape)
print("Total: ",X_train.shape[0]+X_test.shape[0])
```
